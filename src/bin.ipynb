```python
import torch
```

```python
with open("../data/nombres_raw.txt", "r") as file:
    lines = file.read().splitlines()

# TODO
bigrams = []

cleaned_names = []
for line in lines:
    words = []
    for word in line.split():
        if not word.replace(',', '').replace('.', '').isdigit():
            cleaned_names.append(word)

for name in cleaned_names:
    name = "-" + name.lower() + "."
    for i in range(len(name)-1):
        bigrams.append((name[i],name[i+1]))

bigrams
```

```python
char_to_idx = dict()
alphabet = "-" + "abcdefghijklmnñopqrstuvwxyz" + "."

for i in range(len(alphabet)):
    char_to_idx[alphabet[i]] = i

char_to_idx
```

```python
num_chars = len(char_to_idx)
bigram_counts: torch.Tensor = torch.zeros(num_chars, num_chars, dtype=torch.int32)

# Iterate over each bigram and update the count in the tensor
# TODO

for char1, char2 in bigrams:
    if char1 in char_to_idx and char2 in char_to_idx:
        i = char_to_idx[char1]
        j = char_to_idx[char2]
        bigram_counts[i, j] += 1

bigram_counts
```

```python
alphabet = "-abcdefghijklmnopqrstuvwxyz."
column_sum = torch.sum(bigram_counts, dim=0)
column_sum

```

```python
recount_per_letter = dict()
for i in range(len(alphabet)):
    recount_per_letter[alphabet[i]] = column_sum[i].item()

recount_per_letter
```
